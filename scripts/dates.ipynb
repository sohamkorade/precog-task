Calendar heatmaps for cases

Holidays are marked by circles

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cal
```

```python
for year in range(2010, 2018 + 1):
    # read csv
    df = pd.read_csv(f"../csv/cases/cases/small_cases_{year}.csv")

    # convert `date_of_filing` column to datetime
    daywise = pd.to_datetime(df['date_of_filing'])

    # get frequency count
    freq = daywise.value_counts()

    # plot
    cal.plot(freq, year)
```

```python
plt.figure(figsize=(10, 6))

years = range(2010, 2018 + 1)

for year in years:
    # read csv
    df = pd.read_csv(f"../csv/cases/cases/small_cases_{year}.csv")

    # convert `date_of_filing` column to datetime
    monthwise = pd.to_datetime(df['date_of_filing']).dt.month

    # get frequency count
    freq = monthwise.value_counts().sort_index()

    # plot
    plt.plot(freq, label=f"{year}")
plt.xlim(left=1)
plt.ylim(bottom=freq.min() - 100)
plt.xticks(np.arange(1, 12 + 1), cal.month_names, rotation=45)
plt.xlabel("Month")
plt.ylabel("Number of cases")
plt.title("Monthwise cases")
plt.legend(years)
plt.show()
```
