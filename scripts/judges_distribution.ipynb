# Regional distribution
To understand the distribution of judges across different regions of India, we create a map showing the locations of the courts where the judges in our dataset serve. We also calculate statistics, such as the average number of judges per court, to understand regional variations in judicial representation.

```python
import pandas as pd

judges = pd.read_csv("../csv/judges_clean.csv")
```

```python
# group by state_code and dist_code
grouped = judges.groupby(["state_code", "dist_code"])

# plot the distribution of judges
grouped.size().plot(kind="bar", figsize=(20, 10))
```
