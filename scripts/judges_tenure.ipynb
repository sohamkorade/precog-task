patterns in start and end dates of judges' tenure

```python
import pandas as pd

judges = pd.read_csv("../csv/judges_clean.csv")
```

```python
# keep only start_date
df = judges[['start_date']]

# plot graph of start_date after converting to datetime
df["start_date"] = pd.to_datetime(df["start_date"])

df.groupby(df["start_date"].dt.month).count().plot(kind='bar', figsize=(12, 4))
```

```python
# keep only end_date
df = judges[['end_date']]

# plot graph of end_date after converting to datetime
df["end_date"] = pd.to_datetime(df["end_date"])

df.groupby(df["end_date"].dt.month).count().plot(kind='bar', figsize=(12, 4))
```
